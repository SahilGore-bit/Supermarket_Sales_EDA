```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Setup
%matplotlib inline
sns.set(style="whitegrid")

# Load Data
df = pd.read_csv("supermarket_sales.csv")
```

```python
# Display First 5 Rows
print("First 5 Rows of the Dataset:")
display(df.head())

# Dataset Info
print("\nDataset Info:")
df.info()

# Check for Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Basic Stats
print("\nStatistical Summary:")
display(df.describe())

# Column Names
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
#  Exploratory Data Analysis

# 1. Count of Invoices by Branch
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Branch', palette='Set2')
plt.title("Number of Invoices by Branch")
plt.xlabel("Branch")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Invoice_Count_by_Branch.png")
plt.show()

# 2. Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Gender_Distribution.png")
plt.show()

# 3. Product Line vs. Gross Income
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Product line', y='gross income', palette='Set3')
plt.title("Gross Income by Product Line")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Gross_Income_by_Product_Line.png")
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("Correlation_Matrix.png")
plt.show()

# 5. Payment Method Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Payment', palette='Set1')
plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Payment_Method_Distribution.png")
plt.show()
```
